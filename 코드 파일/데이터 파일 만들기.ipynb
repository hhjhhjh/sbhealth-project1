### id별로 데이터셋 생성

```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

# 올바른 파일 경로 확인
# Make sure this path is correct by copying it from your Google Drive in Colab
data_zip_path = '/content/drive/MyDrive/1차 플젝/건강데이터_id.csv'

# CSV 파일 읽기
data2 = pd.read_csv(data_zip_path, encoding='cp949')

# 데이터 확인
data2.head()
```

```python
# 삭제할 컬럼 리스트에 포함시켜 제거
columns_to_drop = ['날짜', '키', '체중']  # 삭제하고 싶은 컬럼명 입력
data2 = data2.drop(columns=columns_to_drop)

# 결과 확인
data2.head()
```

```python
data2['지속 시간'] = data2['지속 시간'].round(2)
data2['칼로리 소모'] = data2['칼로리 소모'].round(2)
data2['평균 심박수'] = data2['평균 심박수'].round(2)
data2['수면 시간'] = data2['수면 시간'].round(2)
data2['스트레스 수준'] = data2['스트레스 수준'].round(2)
data2['매일 걸음수'] = data2['매일 걸음수'].round(2)
data2['체내 수분 함량'] = data2['체내 수분 함량'].round(2)
data2['체력 수준'] = data2['체력 수준'].round(2)

# 결과 출력
print(data2.head())
```

```python
data2.to_csv('/content/drive/MyDrive/1차 플젝/건강데이터_id.csv', index=False, encoding='cp949')
```

```python
from google.colab import files
import pandas as pd

# 데이터 불러오기 및 전처리
file_path = '/content/drive/MyDrive/1차 플젝/건강데이터_키_몸무게_삭제.csv'
data = pd.read_csv(file_path, encoding='cp949')

def categorize_bmi(BMI):
    if BMI < 18.5:
        return '저체중'
    elif 18.5 <= BMI < 23:
        return '정상'
    elif 23 <= BMI < 25:
        return '과체중'
    else:
        return '비만'

data['BMI 분류'] = data['BMI'].apply(categorize_bmi)

data.head()
```

```python
data.to_csv('/content/drive/MyDrive/1차 플젝/건강데이터_bmi_분류.csv', index=False, encoding='cp949')
```
